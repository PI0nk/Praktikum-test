```python
import sys
import re
import numpy as np
import matplotlib.pyplot as plot
import scipy as scp
from scipy import optimize
from scipy import stats
plot.rcParams.update({'font.size': 20})
import pandas as pd
```

```python
f = open("Versuchsteil2.lab")

müll = []
n = []
t = []
E = []


for i in range(22):
    müll.append(f.readline())
for i in range(5581-22):
    s = f.readline()
    s_arr = re.split('\t',s)
    n.append(int(s_arr[0]))
    t.append(float(s_arr[1]))
    E.append(float(s_arr[2]))   
    
```

```python
fig, axs = plot.subplots(nrows=1, ncols=1, figsize=(14, 14), dpi=80)

x_min = 0
x_max = 560
y_min = 0
y_max = 1

ax=axs
#Settings
ax.set_title('Versuchsteil 2')

ax.plot(t,E, label="Messwerte", color="red")

ax.plot([31,31],[0,1],color="darkslategrey")
ax.plot([58,58],[0,1],color="lime")
ax.text(0.09, 0.9, "I",transform=ax.transAxes, fontsize=40,verticalalignment='top')
ax.plot([85,85],[0,1],color="darkslategrey")

ax.plot([117,117],[0,1],color="darkslategrey")
ax.plot([170,170],[0,1],color="lime")
ax.text(0.255, 0.9, "II",transform=ax.transAxes, fontsize=40,verticalalignment='top')
ax.plot([190,190],[0,1],color="darkslategrey")

ax.plot([245,245],[0,1],color="darkslategrey")
ax.plot([255,255],[0,1],color="lime")
ax.text(0.465, 0.9, "III",transform=ax.transAxes, fontsize=40,verticalalignment='top')
ax.plot([301,301],[0,1],color="darkslategrey")

ax.plot([336,336],[0,1],color="darkslategrey")
ax.plot([336,336],[0,1],color="lime")
ax.text(0.61, 0.9, "IV",transform=ax.transAxes, fontsize=40,verticalalignment='top')
ax.plot([377,377],[0,1],color="darkslategrey")

ax.plot([385,385],[0,1],color="darkslategrey")
ax.plot([385,385],[0,1],color="lime")
ax.text(0.73, 0.9, "V",transform=ax.transAxes, fontsize=40,verticalalignment='top')
ax.plot([461,461],[0,1],color="darkslategrey")

ax.plot([505,505],[0,1],color="darkslategrey")
ax.plot([525,525],[0,1],color="lime")
ax.text(0.92, 0.9, "VI",transform=ax.transAxes, fontsize=40,verticalalignment='top')
ax.plot([556,556],[0,1],color="darkslategrey")


ax.set_xlabel("Zeit t in s")
ax.set_ylabel("Beleuchtungsstärke E in W/m^2")

#Grid Setting
major_ticks = np.arange(x_min, x_max, 100)
minor_ticks = np.arange(x_min, x_max, 10)
major_ticks_y = np.arange(y_min, y_max, 0.1)
minor_ticks_y = np.arange(y_min, y_max, 0.01)

ax.set_xlim([x_min,x_max])
ax.set_ylim([y_min,y_max])

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks_y)
ax.set_yticks(minor_ticks_y, minor=True)

ax.grid(which='minor', alpha=0.55)
ax.grid(which='major', alpha=0.8)

ax.grid(which='both')
plot.grid()
plot.legend()
plot.savefig('Matplot.png')
```
